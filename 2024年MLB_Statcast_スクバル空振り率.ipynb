```python
#CSVファイル読み込みと結合(sample_dataフォルダ内)
import glob
```

```python
#ファイル名パターンに一致するCSVを取得
csv_files_pattern="sample_data/statcast_*.csv"
csv_files=glob.glob(csv_files_pattern)
```

```python
#読み込んだDataFrameを格納するリスト
dfs=[]
```

```python
#各CSVファイルを読み込み
for file in csv_files:
  try:
    df_month=pd.read_csv(file)
    dfs.append(df_month)
    print(f"{file}を読み込みました。")
  except Exception as e:
    print(f"{file}の読み込み中にエラーが発生しました: {e}")
```

```python
#全てのデータフレームを結合
if dfs:
  df_combined=pd.concat(dfs, ignore_index=True)
  df=df_combined #メイン変数としてdfに代入
  print("全てのデータフレームを結合しました。")
  display(df.head()) #先頭5行を表示
  print(f"結合後のデータフレームの行数: {len(df)}")
else:
  print("読み込めるCSVファイルが見つかりませんでした。")
```

```python
#PITCHER_IDを設定
pitcher_id=669373
```

```python
#1 対象投手のデータをフィルタリング
#* df_mergedは、データ取得・結合後のDataFrame名としています
df_skubal=df[df["pitcher"]==pitcher_id]
```

```python
# ***【デバッグ用コードの追記】***
print("--- [デバッグ]スクーバル選手が投げた全球種のリスト ---")
print(df_skubal["pitch_name"].unique())
print("-------------------------------------------------------\n")
```

```python
#2 球種ごとの平均速度を計算
avg_velo=df_skubal.groupby("pitch_name")["release_speed"].mean().reset_index()
avg_velo=avg_velo.rename(columns={"release_speed": "avg_velo"})
```

```python
# *** 【デバッグ用コードの追記】***
print("--- [デバッグ] avg_veloの中身 ---")
print(avg_velo)
print("----------------------------------\n")
```

```python
#3 フォーシームとチェンジアップの球速を取得
try:
  velo_fastball=avg_velo[avg_velo["pitch_name"]=="4-Seam Fastball"]["avg_velo"].iloc[0]
  velo_changeup=avg_velo[avg_velo["pitch_name"]=="Changeup"]["avg_velo"].iloc[0]

  #4 速度差(Velo Differential)を計算
  velo_differential=velo_fastball - velo_changeup

  #結果の表示
  print("--- 669373 (タリク・スクーバル)選手の平均速度---")
  print(avg_velo)
  print("\n--- フォーシームとチェンジアップの速度差 ---")
  print(f"フォーシーム平均: {velo_fastball:.2f} mph")
  print(f"チェンジアップ平均: {velo_changeup:.2f} mph")
  print(f" 速度差(Velo Differential): {velo_differential: .2f} mph")
except IndexError:
  print("エラー: データセットに対象の球種(4-Seam Fastball または Changeup)が見つかりません")
  velo_differential=None

#次の考察ステップのために値を保持
skubal_velo_differential=velo_differential
```

```python
#Whiff%を計算する関数を定義
def calculate_whiff_percentage(df):
  #1.スイングが発生した投球をフィルタリング
  #Statcastでは"description"列でスイングの有無を判断する
  #スイングを定義する主な値(Swinging Strike,Foul,Play)
  swing_descriptions=[
      "swinging_strike",
      "swinging_strike_blocked",
      "foul",
      "foul_tip",
      "hit_into_play",
      "missed_bunt",
      "foul_bunt"
  ]
  df_swings=df[df["description"].isin(swing_descriptions)].copy()
  #2 空振り (Swinging Strike)をカウント
  # "swinging_strike"または"swinging_strike_blocked"が三振とみなされる
  whiff_descriptions=["swinging_strike","swinging_strike_blocked"]
  #全投球を対象に 球種ごとの空振り数をカウント
  whiffs_by_pitch = df[df["description"].isin(whiff_descriptions)].groupby("pitch_name").size().reset_index(name="Whiffs")
  #3 総スイング数（Total Swings)をカウント
  #スイングした投球を対象に球種ごとに総スイング数をカウント
  swings_by_pitch=df_swings.groupby("pitch_name").size().reset_index(name="Total_Swings")
  #4.データフレームを結合しWhiff%を計算
  whiff_data=pd.merge(whiffs_by_pitch, swings_by_pitch, on="pitch_name",how="inner")
  #Whiff%=(Whiffs/Total_Swings)*100
  whiff_data["Whiff_Percentage"]=(whiff_data["Whiffs"]/whiff_data["Total_Swings"])*100
  #精度を上げるため小数点第2位まで表示
  whiff_data["Whiff_Percentage"]=whiff_data["Whiff_Percentage"].round(2)
  return whiff_data
```

```python
#Whiff%を計算
skubal_whiff_results=calculate_whiff_percentage(df_skubal)
#結果の表示
print("---669373(タリク・スクーバル)選手の球種別Whiff%")
print(skubal_whiff_results)
print("-----------------------------------------------")
```
