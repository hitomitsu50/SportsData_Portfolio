```python
#2025年10月21日
#データの読み込み
df=pd.read_csv("sample_data/2022_audience.csv")
df
```

```python
#本拠地のみ抽出
df=pd.read_csv("sample_data/2022_audience.csv")
df_home=df[df["本拠地"]==True]
df_home["球場"].value_counts()
```

```python
#球場名をリストに格納
stadiums=["PayPayドーム","ベルーナドーム","楽天生命パーク","京セラドーム","ZOZOマリン","札幌ドーム",
         "横浜スタジアム","東京ドーム","神宮","マツダスタジアム","バンテリンドーム","甲子園"]
```

```python
#結果を格納する空リストを作成
stadium_average=[]
#球場名を順番に代入
for i, stadium in enumerate(stadiums):
  stadium_data=df_home[df_home["球場"]==stadium]
  if i == 4: # 5回目のループ (インデックスは4)
    display(stadium_data)
  average=stadium_data["観客動員数"].mean()
  stadium_average.append(average)
stadium_average
```

```python
#棒グラフの作成
plt.figure(figsize=(20,8))
plt.bar(stadiums,stadium_average)
plt.title("球場別の観客動員数")
plt.show()
```

```python
#2025年10月22日
#曜日名をリストに格納
days=["月","火","水","木","金","土","日"]
```

```python
#結果を格納する空リストを作成
day_average=[]
#曜日名を順番に代入
for day in days:
  day_data=df_home[df_home["曜日"]==day]
  average=day_data["観客動員数"].mean()
  day_average.append(average)
day_average
```

```python
#折れ線グラフの作成
plt.figure(figsize=(10,6))
plt.plot(days,day_average)
plt.title("曜日別の観客動員数")
plt.xlabel("曜日")
plt.ylabel("観客動員数 (平均)")
plt.grid(True)
plt.show()
```

```python
#セリーグのチーム名の頭文字をリストに格納
central_teams=["阪","広","D","巨","ヤ","中"]
```

```python
#結果を格納する空リストを作成
central_numbers=[]
#homeの中にチーム名を順番に代入
for home in central_teams:
  home_data=df[df["Hチーム"]==home]
  lst=[]
  #visitorの中にチーム名を順番に代入
  for visitor in central_teams:
    home_visitor_data=home_data[home_data["Vチーム"]==visitor]
    average=home_visitor_data["観客動員数"].mean()
    lst.append(average)
  central_numbers.append(lst)
central_numbers
```

```python
#ヒートマップ作成
plt.figure(figsize=(8,8))
plt.imshow(central_numbers,cmap="OrRd")

#見た目の調整
plt.colorbar()
plt.title("対戦カード別の平均観客動員数（セリーグ）")
plt.xticks([0,1,2,3,4,5],central_teams)
plt.yticks([0,1,2,3,4,5],central_teams)
plt.xlabel("ビジターチーム")
plt.ylabel("ホームチーム")
plt.show()
```

```python
#2025年10月23日
#SNSに関する分析
#データの読み込み
df=pd.read_csv("sample_data/X_data.csv")
df
```

```python
#折れ線グラフの作成
plt.plot(df["インプレッション"])
plt.title("インプレッションの推移")
plt.xlabel("ポスト数")
plt.ylabel("インプレッション数")
plt.ylim(0,5000)
plt.show()
```

```python
#折れ線グラフの作成
plt.plot(df["エンゲージメント"])
plt.title("エンゲージメント数の推移")
plt.xlabel("ポスト数")
plt.ylabel("エンゲージメント数")
plt.ylim(0,2000)
plt.show()
```

```python
#散布図の作成
plt.scatter(df["インプレッション"],df["エンゲージメント"],alpha=0.5)

#見た目の調整
plt.title("インプレッションとエンゲージメントの関係")
plt.xlabel("インプレッション")
plt.ylabel("エンゲージメント")
plt.xlim(0,1000)
plt.ylim(0,500)
plt.show()
```

```python
#データ格納用のリストを作成
x,y=[],[]
#0時台から23時までのツイートを順に抽出し、インプレッション数の平均をyに格納
for i in range(24):
  x.append(i)
  time_sweet=df[df["時"]==i]
  y.append(time_sweet["インプレッション"].mean())
#棒グラフの作成
plt.bar(x,y)
plt.title("時間帯別の平均インプレッション数")
plt.xlabel("時間帯")
plt.ylabel("平均インプレッション数")
plt.show()
```

```python
#データ格納用のリストを作成
x,y=[],[]
#0時台かた23時台までのポストを順に抽出し、各時間帯の総ポスト数をyに格納
for i in range(24):
  x.append(i)
  time_sweet=df[df["時"]==i]
  y.append(len(time_sweet))

#棒グラフの作成
plt.bar(x,y)
plt.title("時間帯別のポスト数")
plt.xlabel("時間帯")
plt.ylabel("ポスト数")
plt.show()
```
