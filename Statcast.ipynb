```python
from pybaseball import statcast
import pandas as pd
#2024年4月1日～7日 statcastデータを取得
df=statcast(start_dt="2024-04-01",end_dt="2024-04-07")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib

# Now generate the plot in the same cell
plt.figure(figsize=(10,6))
df_reset = df.reset_index(drop=True) # Reset the index
df_reset['pitch_type'] = df_reset['pitch_type'].astype('category')
sns.countplot(data=df_reset,x="pitch_type",order=df_reset["pitch_type"].value_counts().index)
plt.title("2024年4月1日～7日 投球分布")
plt.xlabel("球種")
plt.ylabel("投球数")
plt.grid(True)
plt.show()
```

```python
#関数化
def plot_pitch_distribution(start_date,end_date):
  df=statcast(start_dt=start_date,end_dt=end_date)
  plt.figure(figsize=(10,6))
  sns.countplot(data=df,x="pitch_type",order=df["pitch_type"].value_counts().index)
  plt.title(" 投球分布")
  plt.xlabel("球種")
  plt.ylabel("投球数")
  plt.xticks(rotation=45)
  plt.grid(True)
  plt.show()
```
